```python
import pandas as pd

# Update path if needed
df = pd.read_csv('/content/drive/MyDrive/CVD Dataset.csv')
df.head()
```

```python
df.drop('Height (m)', axis=1, inplace=True)
```

```python
df.drop('Waist-to-Height Ratio', axis=1, inplace=True)
```

```python
df = df[df['Estimated LDL (mg/dL)'] > 0]
```

```python
df.fillna(df.median(numeric_only=True), inplace=True)
```

```python
df.head()
```

```python
df.isnull().sum()
```

```python
df = df[(df['Systolic BP'] > 90) & (df['Systolic BP'] < 200)]
df = df[(df['Diastolic BP'] > 60) & (df['Diastolic BP'] < 130)]
```

```python
df = df[(df['Total Cholesterol (mg/dL)'] > 100) & (df['Total Cholesterol (mg/dL)'] < 300)]
```

```python
df.shape
```

```python
df = df[(df['Fasting Blood Sugar (mg/dL)'] > 70) & (df['Fasting Blood Sugar (mg/dL)'] < 200)]
```

```python
df['pulse_pressure'] = df['Systolic BP'] - df['Diastolic BP']
```

```python
df['chol_ratio'] = df['Total Cholesterol (mg/dL)'] / df['HDL (mg/dL)']
```

```python
df['target'] = (df['CVD Risk Score'] > 17).astype(int)
```

```python
df.head()
```

```python
df.drop('Blood Pressure (mmHg)', axis=1, inplace=True)
```

```python
df['Sex'] = df['Sex'].map({'M': 1, 'F': 0})
```

```python
df['Smoking Status'] = df['Smoking Status'].map({'Y': 1, 'N': 0})
```

```python
bp_map = {
    'Normal': 0,
    'Elevated': 1,
    'Hypertension Stage 1': 2,
    'Hypertension Stage 2': 3
}

df['BP_Category'] = df['Blood Pressure Category'].map(bp_map)
df.drop('Blood Pressure Category', axis=1, inplace=True)
```

```python
df.drop('CVD Risk Level', axis=1, inplace=True)
```

```python
X = df.drop('target', axis=1)
y = df['target']
```

```python
import pandas as pd
import numpy as np

from sklearn.model_selection import train_test_split
from sklearn.preprocessing import StandardScaler
from xgboost import XGBClassifier
from sklearn.metrics import accuracy_score, roc_auc_score

# ----------------------------
# FEATURES & TARGET
# ----------------------------
X = df.drop(['target'], axis=1)
y = df['target']

# ----------------------------
# ONE-HOT ENCODING for remaining categorical features
# ----------------------------
categorical_cols = X.select_dtypes(include=['object']).columns
X = pd.get_dummies(X, columns=categorical_cols, drop_first=True)

# ----------------------------
# SCALING
# ----------------------------
scaler = StandardScaler()
X_scaled = scaler.fit_transform(X)

# ----------------------------
# SPLIT
# ----------------------------
X_train, X_test, y_train, y_test = train_test_split(
    X_scaled, y, test_size=0.2, random_state=42
)

# ----------------------------
# MODEL
# ----------------------------
model = XGBClassifier(n_estimators=300, max_depth=5)
model.fit(X_train, y_train)

# ----------------------------
# EVALUATION
# ----------------------------
y_pred = model.predict(X_test)
y_prob = model.predict_proba(X_test)

print("Accuracy:", accuracy_score(y_test, y_pred))
print("ROC AUC:", roc_auc_score(y_test, y_prob[:,1]))
```

```python
import numpy as np
import pandas as pd
import joblib

# ---------------- LOAD EVERYTHING ----------------
model = joblib.load('/content/model.pkl')
scaler = joblib.load('/content/scaler.pkl')
features = joblib.load('/content/features.pkl')  # saved earlier

# ---------------- USER INPUT ----------------
print("===== Enter Patient Details =====")

sex = int(input("Sex (Male=1, Female=0): "))
age = float(input("Age (years): "))
weight = float(input("Weight (kg): "))
bmi = float(input("BMI: "))
ab_circ = float(input("Abdominal Circumference (cm): "))
total_chol = float(input("Total Cholesterol (mg/dL): "))
hdl = float(input("HDL: "))
fasting_bs = float(input("Fasting Blood Sugar: "))
smoking = int(input("Smoking (Yes=1, No=0): "))
diabetes = int(input("Diabetes (Yes=1, No=0): "))
family_history = int(input("Family History (Yes=1, No=0): "))
height = float(input("Height (cm): "))
systolic_bp = float(input("Systolic BP: "))
diastolic_bp = float(input("Diastolic BP: "))
ldl = float(input("Estimated LDL: "))

# ---------------- DERIVED FEATURES ----------------
pulse_pressure = systolic_bp - diastolic_bp
chol_ratio = total_chol / hdl if hdl != 0 else 0

# BP Category
if systolic_bp < 120:
    bp_cat = 0
elif systolic_bp < 130:
    bp_cat = 1
elif systolic_bp < 140:
    bp_cat = 2
else:
    bp_cat = 3

# ---------------- CREATE INPUT DICTIONARY ----------------
input_dict = {
    'Sex': sex,
    'Age': age,
    'Weight (kg)': weight,
    'BMI': bmi,
    'Abdominal Circumference (cm)': ab_circ,
    'Total Cholesterol (mg/dL)': total_chol,
    'HDL (mg/dL)': hdl,
    'Fasting Blood Sugar (mg/dL)': fasting_bs,
    'Smoking Status': smoking,
    'Diabetes Status': diabetes,
    'Family History of CVD': family_history,
    'Height (cm)': height,
    'Systolic BP': systolic_bp,
    'Diastolic BP': diastolic_bp,
    'Estimated LDL (mg/dL)': ldl,
    'pulse_pressure': pulse_pressure,
    'chol_ratio': chol_ratio,
    'BP_Category': bp_cat
}

# ---------------- CONVERT TO DATAFRAME ----------------
user_df = pd.DataFrame([input_dict])

# ---------------- MATCH TRAINING FEATURES ----------------
for col in features:
    if col not in user_df.columns:
        user_df[col] = 0  # fill missing columns

# Ensure correct order
user_df = user_df[features]

# ---------------- SCALE ----------------
user_scaled = scaler.transform(user_df)

# ---------------- PREDICT ----------------
prediction = model.predict(user_scaled)
prob = model.predict_proba(user_scaled)

# ---------------- OUTPUT ----------------
print("\n===== RESULT =====")
print(f"Heart Risk Probability: {prob[0][1]*100:.2f}%")

if prediction[0] == 1:
    print("⚠️ High Risk of Heart Disease")
else:
    print("✅ Low Risk")
```
